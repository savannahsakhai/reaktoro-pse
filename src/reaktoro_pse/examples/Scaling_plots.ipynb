```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
import matplotlib


from analysis_plot_kit.core import fig_generator
```

```python
data = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\reaktoro-pse\data_scaling_reaktoro.csv')
```

```python
sea_color = "#FFD046"
nacl_color = "#4A7B9D"
```

```python
vmin = 0
vmax = 1

zticks = np.linspace(vmin, vmax, 5)
zticks = np.around(zticks, decimals=2)

TDSticks = [3.361113e-02,4.838276e-02,9.373868e-02,1.366460e-01,1.775197e-01]
tempticks = np.linspace(25 + 273.15, 95 + 273.15, 8)
pHticks = np.linspace(5.5, 10.5, 6)
pticks = [1e4, 5e4, 1e5, 4e6, 8e6]

```

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(3.5*3, 2.5*3)) #figsize=(3.5, 2.5)

axs[0,0].plot(data['# Feed TDS'], data['ST Calcite'], color=sea_color)
axs[0,0].plot(data['# Feed TDS'], data['ST Gypsum'], color=nacl_color)
axs[0,0].set_xlabel("Feed TDS kg/s")
axs[0,0].set_ylabel("ST")

axs[0,1].plot(data['Temperature'], data['ST Calcite'], color=sea_color)
axs[0,1].plot(data['Temperature'], data['ST Gypsum'], color=nacl_color)
axs[0,1].set_xlabel("Temperature")
axs[0,1].set_ylabel("ST")

axs[1,0].plot(data['pH'], data['ST Calcite'], color=sea_color)
axs[1,0].plot(data['pH'], data['ST Gypsum'], color=nacl_color)
axs[1,0].set_xlabel("pH")
axs[1,0].set_ylabel("ST")

axs[1,1].plot(data['Pressure']/1e5, data['ST Calcite'], color=sea_color)
axs[1,1].plot(data['Pressure']/1e5, data['ST Gypsum'], color=nacl_color)
axs[1,1].set_xlabel("Pressure")
axs[1,1].set_ylabel("ST")

axs[1,0].legend(["Calcite","Gypsum"],title="Mineral", loc="lower right")
```

```python
def st_cmap(data,x,y,xticks,yticksz,z,color="GnBu"):
    figure = fig_generator.figureGenerator()
    figure.init_figure()
    figure.colorMaps["color_map"] = cm.get_cmap(color)
    figure.plot_map(
        xdata=data[x],
        ydata=data[y],
        zdata=data[z],
        text=False,
        build_map=True,
        vmin=vmin,
        vmax=vmax,
    )
    figure.add_colorbar(zticks=zticks, zformat=2, zlabel="Scaling Tendency")
    xticks = np.linspace(min(data[x]), max(data[x]), 5)
    yticks = np.linspace(min(data[y]), max(data[y]), 5)
    figure.set_axis_ticklabels(
        xticklabels=xticks,
        yticklabels=yticks,
        xformat=3,
        yformat=3,
        xlabel=x,
        ylabel=y,
    )
```

```python
st_cmap(data,"# Feed TDS","Temperature",TDSticks,tempticks,"ST Calcite")
st_cmap(data,"# Feed TDS","pH",TDSticks,pHticks,"ST Calcite")
st_cmap(data,"# Feed TDS","Pressure",TDSticks,pticks,"ST Calcite")

st_cmap(data,"Temperature","pH",tempticks,pHticks,"ST Calcite")
st_cmap(data,"Temperature","Pressure",tempticks,pticks,"ST Calcite")

st_cmap(data,"pH","Pressure",pHticks,pticks,"ST Calcite")
```
